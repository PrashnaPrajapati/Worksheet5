 Step-2- Build a Cost Function:

```python
def cost_function(X, Y, W):
    """
    Calculate the Mean Squared Error (MSE).

    Parameters:
    X: Feature matrix (d x n)
    Y: Target matrix (1 x n)
    W: Weight matrix (d x 1)

    Returns:
    cost: Mean squared error
    """
    # Predicted values: Y_pred = W^T X
    Y_pred = np.dot(W.T, X)

    # Difference between predicted and actual values
    loss = Y_pred - Y

    # MSE formula: (1 / (2n)) * sum((Y_pred - Y)^2)
    cost = (1 / (2 * X.shape[1])) * np.sum(np.square(loss))
    return cost

#Test case
X_test = np.array([[1, 3, 5], [2, 4, 6]])
Y_test = np.array([[3, 7, 11]])
W_test = np.array([[1], [1]])

cost = cost_function(X_test, Y_test, W_test)
print("Cost:", cost)
```

Step-3- Gradient Descent for Simple Linear Regression:

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

def gradient_descent(X, Y, W, alpha, iterations):
    m = X.shape[1]
    cost_history = []

    for i in range(iterations):
        # Compute the model's prediction
        predictions = np.dot(W.T, X)

        # Calculate the loss (error)
        loss = predictions - Y

        # Calculate the cost (mean squared error)
        cost = (1 / (2 * m)) * np.sum(np.square(loss))
        cost_history.append(cost)

        # Compute the gradient of the cost function
        gradients = (1 / m) * np.dot(X, loss.T)

        # Update the weights
        W = W - alpha * gradients

        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost}")

    return W, cost_history

def rmse(Y_true, Y_pred):
    return np.sqrt(np.mean((Y_true - Y_pred) ** 2))

def r2(Y_true, Y_pred):
    ss_total = np.sum((Y_true - np.mean(Y_true)) ** 2)
    ss_residual = np.sum((Y_true - Y_pred) ** 2)
    return 1 - (ss_residual / ss_total)
```

```python
def rmse(Y, Y_pred):
    """
    Calculate the Root Mean Squared Error (RMSE).
    """
    mse = np.mean(np.square(Y_pred - Y))
    return np.sqrt(mse)

def r2(Y, Y_pred):
    """
    Calculate the Coefficient of Determination (R²).
    """
    ss_total = np.sum(np.square(Y - np.mean(Y)))
    ss_residual = np.sum(np.square(Y - Y_pred))
    return 1 - (ss_residual / ss_total)

# Example usage
Y_pred_test = np.dot(W_optimal.T, X_test)  # Predicted values
print("\nRMSE:", rmse(Y_test, Y_pred_test))
print("R²:", r2(Y_test, Y_pred_test))
```

 3.1.5 Step - 5  - Main Function to Integrate All Steps:

 We will define a function that:
 1. Loads the data and splits it into training and test sets.

```python
def main():
    # Step 1: Load the dataset
    data = pd.read_csv('/content/drive/MyDrive/Datasets-AI/student.csv')

    # Step 2: Prepare feature matrix (X) and target vector (Y)
    X = data[['Math', 'Reading']].values.T  # d x n
    Y = data['Writing'].values.reshape(1, -1)  # 1 x n

    # Step 3: Split the dataset into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X.T, Y.T, test_size=0.2, random_state=42)
    X_train, X_test, Y_train, Y_test = X_train.T, X_test.T, Y_train.T, Y_test.T  # Transpose back

    # Step 4: Scale the features (Math, Reading)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train.T).T  # Scale training data
    X_test = scaler.transform(X_test.T).T  # Scale test data

    # Step 5: Initialize weights (W), learning rate, and iterations
    W = np.zeros((X_train.shape[0], 1))  # d x 1
    alpha = 0.001  # Lower learning rate
    iterations = 1000

    # Step 6: Perform gradient descent
    W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    # Step 7: Make predictions on the test set
    Y_pred_test = np.dot(W_optimal.T, X_test)

    # Step 8: Evaluate the model
    model_rmse = rmse(Y_test, Y_pred_test)
    model_r2 = r2(Y_test, Y_pred_test)

    # Step 9: Output results
    print("\nOptimized Weights (W):", W_optimal)
    print("\nRMSE on Test Set:", model_rmse)
    print("R² on Test Set:", model_r2)
    print("\nFirst 10 Costs:", cost_history[:10])

# Run the main function
if __name__ == "__main__":
    main()
```
